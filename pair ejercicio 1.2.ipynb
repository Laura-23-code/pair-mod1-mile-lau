5. Verificar si una letra es una vocal o una consonante

Descripción: Dada una letra, verifica si es una vocal o una consonante.

```python
letra = input("Introduce una letra:")
if letra in ['a', 'e', 'i', 'o', 'u']:
    print("Es una vocal")
else:
    print("Es una consonante")
```
